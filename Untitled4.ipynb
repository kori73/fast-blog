```python
from scipy.stats import poisson, norm
```

```python
rv = poisson(mu=5)
```

```python
import numpy as np
x = np.arange(0, 50, 0.01)
probs = rv.pmf(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x, probs)
```

```python
mean = (x * probs).sum()
```

```python
mean
```

```python
((x - mean)** 2 * probs).sum()
```

```python
y = np.sqrt(x)
```

```python
plt.bar(y, probs)
```

```python
normal = norm(1, 1.25)
```

```python
vals = np.arange(-5, 5, 0.1)
```

```python
normal_pdf = normal.pdf(vals)
```

```python
plt.plot(vals, normal_pdf)
for i in vals:
    plt.axvline(i)
```

```python
def logistic(x):
    return 1 / (1 + np.exp(-x))
```

```python
transformed = logistic(vals)
```

```python
for i in transformed:
    plt.axvline(i)
```

```python
def inv(y):
    return np.log(1 / (1 - y))
```

```python
def abs_j(y):
    return 1 / (y* (1- y))
```

```python
plt.plot(transformed, normal_pdf * abs_j(transformed))
for i in transformed:
    plt.axvline(i)
```
